**Journals & Ranking**

An academic journal or research journal is a periodical publication in which research articles relating to a particular academic discipline is published, according to Wikipedia. Currently, there are more than 25,000 peer-reviewed journals that are indexed in citation index databases such as Scopus and Web of Science. These indexes are ranked on the basis of various metrics such as CiteScore, H-index, etc. The metrics are calculated from yearly citation data of the journal. A lot of efforts are given to make a metric that reflects the journal's quality.

**Key Features**

**Rank**: Overall rank of journal (derived from sorted SJR index).

**Title**: Name or title of journal.

**OA**: Open Access or not.

**Country**: Country of origin.

**SJR-index**: A citation index calculated by Scimago.

**CiteScore**: A citation index calculated by Scopus.

**H-index**: Hirsh index, the largest number h such that at least h articles in that journal were cited at least h times each.

Best Quartile: Top Q-index or quartile a journal has in any subject area.

Best Categories: Subject areas with top quartile.

Best Subject Area: Highest ranking subject area.

Best Subject Rank: Rank of the highest ranking subject area.

Total Docs.: Total number of documents of the journal.

Total Docs. 3y: Total number of documents in the past 3 years.

Total Refs.: Total number of references of the journal.

Total Cites 3y: Total number of citations in the past 3 years.

Citable Docs. 3y: Total number of citable documents in the past 3 years.

Cites/Doc. 2y: Total number of citations divided by the total number of documents in the past 2 years.

Refs./Doc.: Total number of references divided by the total number of documents.

Publisher: Name of the publisher company of the journal.

Core Collection: Web of Science core collection name.

Coverage: Starting year of coverage.

Active: Active or inactive.

In-Press: Articles in press or not.

ISO Language Code: Three-letter ISO 639 code for language.

## **Step 1 - Data Overview and Cleaning**

**Check for data types, identify and handle missing values, and prepare the dataset for a comprehensive analysis.**

```python
df.isnull().sum()
```

```python
# handling the missing values
df['CiteScore'].fillna(df['CiteScore'].median(), inplace=True)
df['Publisher'].fillna('Unknown', inplace=True)
df['Core Collection'].fillna('Unknown', inplace=True)
df['Active'].fillna('Unknown', inplace=True)
df['In-Press'].fillna('Unknown', inplace=True)
df['ISO Language Code'].fillna('Unknown', inplace=True)
df['ASJC Codes'].fillna('Unknown', inplace=True)
df['Coverage'] = pd.to_numeric(df['Coverage'], errors='coerce')
df['Coverage'].fillna(df['Coverage'].median(), inplace=True)

print("Missing values after cleaning:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

## **Step 4 - Geographic and Publisher Influence**

To understand the influence of publishers, let's first identify the top 10 publishers and country with the most journals in the dataset.

```python
top_publishers = df['Publisher'].value_counts().head(10)
top_countries = df['Country'].value_counts().head(10)
print("Top 10 Publishers:")
print(top_publishers)
print("\nTop 10 Countries:")
print(top_countries)
```

As observed in the data the we can see that "Unknown" value is coming in the top position, as they can skew our analysis, we would exclude them. We focus on actual publishing entities.

```python
top_publishers = df[df['Publisher'] != 'Unknown']['Publisher'].value_counts().head(10)
print("Recalculated Top 10 Publishers (excluding 'Unknown'):")
print(top_publishers)
```

Now, let's calculate the average 'SJR-index', 'CiteScore', and 'H-index' for these top 10 publishers.

```python
top_publishers_list = top_publishers.index.tolist()
top_countries_list = top_countries.index.tolist()
average_metrics_by_publisher = df[df['Publisher'].isin(top_publishers_list)].groupby('Publisher')[['SJR-index', 'CiteScore', 'H-index']].mean().sort_values(by='SJR-index', ascending=False)
average_metrics_by_country = df[df['Country'].isin(top_countries_list)].groupby('Country')[['SJR-index', 'CiteScore', 'H-index']].mean().sort_values(by='SJR-index', ascending=False)
print(average_metrics_by_publisher)
print(average_metrics_by_country)
```

Next, let's visualize these average metrics for the top publishers and top countries using bar charts.

## Average Ranking Metrics by Top 10 Countries.

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# Average SJR-index by Countries.
sns.barplot(x=average_metrics_by_country.index, y='SJR-index', data=average_metrics_by_country.sort_values(by='SJR-index', ascending=False), ax=axes[0], palette='viridis', hue=average_metrics_by_country.index, legend=False)
axes[0].set_title('Average SJR-index by Country (Top 10 Countries by Journal Count)')
axes[0].set_xlabel('Country')
axes[0].set_ylabel('Average SJR-index')
axes[0].tick_params(axis='x', rotation=45)

# Average CiteScore by Countries.
sns.barplot(x=average_metrics_by_country.index, y='CiteScore', data=average_metrics_by_country.sort_values(by='CiteScore', ascending=False), ax=axes[1], palette='magma', hue=average_metrics_by_country.index, legend=False)
axes[1].set_title('Average CiteScore by Country (Top 10 Countries by Journal Count)')
axes[1].set_xlabel('Country')
axes[1].set_ylabel('Average CiteScore')
axes[1].tick_params(axis='x', rotation=45)

# Average H-index by Countries.
h_index_sorted_countries = average_metrics_by_country.sort_values(by='H-index', ascending=False)
sns.barplot(x=h_index_sorted_countries.index, y='H-index', data=h_index_sorted_countries, ax=axes[2], palette='cividis', hue=h_index_sorted_countries.index, legend=False)
axes[2].set_title('Average H-index by Country (Top 10 Countries by Journal Count)')
axes[2].set_xlabel('Country')
axes[2].set_ylabel('Average H-index')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# Average SJR-index by Publisher.
sns.barplot(x=average_metrics_by_publisher.index, y='SJR-index', data=average_metrics_by_publisher.sort_values(by='SJR-index', ascending=False), ax=axes[0], palette='viridis', hue=average_metrics_by_publisher.index, legend=False)
axes[0].set_title('Average SJR-index by Publisher (Top 10 Publishers by Journal Count)')
axes[0].set_xlabel('Publisher')
axes[0].set_ylabel('Average SJR-index')
axes[0].tick_params(axis='x', rotation=45)

# Average CiteScore by Publisher.
sns.barplot(x=average_metrics_by_publisher.index, y='CiteScore', data=average_metrics_by_publisher.sort_values(by='CiteScore', ascending=False), ax=axes[1], palette='magma', hue=average_metrics_by_publisher.index, legend=False)
axes[1].set_title('Average CiteScore by Publisher (Top 10 Publishers by Journal Count)')
axes[1].set_xlabel('Publisher')
axes[1].set_ylabel('Average CiteScore')
axes[1].tick_params(axis='x', rotation=45)

# Average H-index by Publisher.
sns.barplot(x=average_metrics_by_publisher.index, y='H-index', data=average_metrics_by_publisher.sort_values(by='H-index', ascending=False), ax=axes[2], palette='cividis', hue=average_metrics_by_publisher.index, legend=False)
axes[2].set_title('Average H-index by Publisher (Top 10 Publishers by Journal Count)')
axes[2].set_xlabel('Publisher')
axes[2].set_ylabel('Average H-index')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Analysis of top-level subject areas among high-ranked journals

top_level_categories = ['Life Sciences','Social Sciences', 'Physical Sciences', 'Health Sciences']

# calculate the count of highly-ranked -journals in each top level category.

top_level_Q1_counts = highly_ranked_journals[top_level_categories].sum()

print("\n Number of Highly-Ranked-Journals in Top-Level Categoreis")
display(top_level_Q1_counts)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_level_Q1_counts.index, y=top_level_Q1_counts.values, palette = 'viridis')
plt.title('Distribution of Highly-Ranked-Journals Across Top Level Category')
plt.xlabel('Top-Level Subject Category')
plt.ylabel('Number of Q1 Journals')
plt.xticks(rotation=0, ha='center')
plt.tight_layout()
plt.show()
```

**Lets also try with average metrics for these top-level categories within the Q1 journals**

```python
# Each category column has 1 if journal belongs to that category, else 0.
# Melt the dataframe to long format: category + metrics
metrics = ['SJR-index', 'CiteScore', 'H-index']

top_level_melted = highly_ranked_journals.melt(
    id_vars=metrics,
    value_vars=top_level_categories,
    var_name='Category',
    value_name='Quartile'
)
print('\nMelted dataframe of metrics across categories')
display(top_level_melted.head())

# Filter only Q1 journals (Quartile ==1)
print('\n')
print('\nFitler only where quartile is Q1')
filtered = top_level_melted[top_level_melted['Quartile']==1]
display(filtered.head())
```

```python
# Group the categories and compute mean metrics


top_level_q1_avg_metrics = filtered.groupby('Category')[metrics].mean().sort_index()
display(top_level_q1_avg_metrics)

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Bar plot for SJR-index for Q1 journals by top-level category
sns.barplot(x = top_level_q1_avg_metrics.index, y='SJR-index', data = top_level_q1_avg_metrics.sort_values(by='SJR-index', ascending=False), ax=axes[0], palette='viridis')
axes[0].set_title('Average SJR-Index by Top-Level Category')
axes[0].set_xlabel('Top-Level Subject Category')
axes[0].set_ylabel('Average SJR-Index')
axes[0].tick_params(axis='x')

# Bar plot for CiteScore for Q1 journals by top-level category
sns.barplot(x = top_level_q1_avg_metrics.index, y='CiteScore', data = top_level_q1_avg_metrics.sort_values(by='CiteScore', ascending=False), ax=axes[1], palette='magma')
axes[1].set_title('Average CiteScore by Top-Level Category')
axes[1].set_xlabel('Top-Level Subject Category')
axes[1].set_ylabel('Average CiteScore')
axes[1].tick_params(axis='x')

# Bar plot for H-index for Q1 journals by top-level category
sns.barplot(x = top_level_q1_avg_metrics.index, y='H-index', data = top_level_q1_avg_metrics.sort_values(by='H-index', ascending=False), ax=axes[2], palette='cividis')
axes[2].set_title('Average H-index by Top-Level Category')
axes[2].set_xlabel('Top-Level Subject Category')
axes[2].set_ylabel('Average H-index')
axes[2].tick_params(axis='x')

plt.tight_layout()
plt.show()
```

## **Step 7 - Correlation Between Metrics**

**Calculate and visualize the correlation matrix of numerical ranking metrics (e.g., 'SJR-index', 'CiteScore', 'H-index', 'Total Docs.', 'Total Cites 3y') to understand the relationships and dependencies between them.**

```python
corr_metrics = df[['SJR-index', 'CiteScore', 'H-index', 'Total Docs.', 'Total Cites 3y']].corr()
display(corr_metrics)
print('\n')
plt.figure(figsize=(8,6))
sns.heatmap(corr_metrics, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Journal Metrics")
plt.show()

```

From the correlation heatmap, we observed a strong positive correlation between 'SJR-index' and 'CiteScore'(0.86). Let's visualize this relationship with a scatter plot.

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='SJR-index', y='CiteScore', data=df, alpha=0.6)
plt.title('Scatter Plot of SJR-index vs. CiteScore')
plt.xlabel('SJR-index')
plt.ylabel('CiteScore')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### **Final Step - Summarize the key findings and insights from the entire analysis.**

### Summary of Key Findings and Insights

Our analysis of journal ranking data has revealed several key insights:

1.  **Data Quality and Preparation:** We started by loading the dataset and handling missing values, particularly for 'CiteScore', 'Publisher', 'Core Collection', 'Coverage', 'Active', 'In-Press', 'ISO Language Code', and 'ASJC Codes'. Missing 'CiteScore' values were imputed with the median, while categorical columns were filled with 'Unknown'.

2.  **Distribution of Ranking Metrics:** The distributions of 'SJR-index', 'CiteScore', and 'H-index' showed a right-skewed pattern, indicating that a majority of journals have lower scores, with a few highly influential journals achieving significantly higher metrics. Box plots further emphasized the presence of outliers, which are common in citation-based metrics.

3.  **Top Journals:** Identification of top journals across 'SJR-index', 'CiteScore', and 'H-index' consistently highlighted well-known publications like "Ca-A Cancer Journal for Clinicians," "Nature Reviews Molecular Cell Biology," "New England Journal of Medicine," and "Nature." This suggests that these metrics often align in identifying highly reputable journals.

4.  **Geographic and Publisher Influence:**
    *   **Publishers:** After excluding 'Unknown' publishers, Wiley, Springer, and Routledge Journals emerged as the top publishers by journal count. In terms of average metric performance, Elsevier, Wiley, and Oxford University Press led in 'SJR-index', 'CiteScore', and 'H-index' among the top publishers, indicating their strong portfolio of high-impact journals.
    *   **Countries:** The United Kingdom and United States dominate in terms of the number of journals. When looking at average metrics, the United States, United Kingdom, and Netherlands generally had the highest average 'SJR-index', 'CiteScore', and 'H-index', suggesting a strong presence of high-quality research output from these regions.

5.  **Open Access vs. Traditional Publishing:** We observed that non-Open Access (Non-OA) journals generally have higher average 'SJR-index', 'CiteScore', and 'H-index' compared to Open Access (OA) journals. This suggests that, on average, traditional publishing models currently host journals with higher impact factors within this dataset.

6.  **Subject Area Performance:**
    *   **Best Subject Area:** "Social Sciences," "Arts and Humanities," and "Medicine" were the most represented 'Best Subject Areas' among Q1 (highly-ranked) journals.
    *   **Top-Level Categories:** "Social Sciences," "Physical Sciences," and "Health Sciences" had the highest number of Q1 journals. When considering average metrics within Q1 journals, "Life Sciences" generally exhibited the highest average 'SJR-index', 'CiteScore', and 'H-index', followed closely by "Health Sciences" and "Physical Sciences." "Social Sciences" had comparatively lower average metrics among the top-level categories, even though it had a high count of Q1 journals.

7.  **Correlation Between Metrics:** A correlation heatmap revealed strong positive correlations between several metrics. Notably, a very strong positive correlation of **0.86** was found between 'SJR-index' and 'CiteScore', indicating that these two metrics largely measure similar aspects of journal influence. 'Total Docs.' and 'Total Cites 3y' also showed a strong positive correlation (0.81). The scatter plot between 'SJR-index' and 'CiteScore' visually confirmed this strong relationship, showing a clear upward trend.
